Αφού καταφέρατε να φθάσετε μέχρι εδώ, θα μπορέσετε να εφαρμόσετε όσα μάθατε στα προηγούμενα μαθήματα ώστε να εμπεδώσετε τις γνώσεις σας και να μπορέσετε να δημιουργήσετε κάτι μόνοι σας.

Ένας φίλος σας έμαθε ότι μόλις τελειώσατε το μάθημα εισαγωγής στον προγραμματισμό με τη Java και σας αναθέτει να του γράψετε το ακόλουθο πρόγραμμα. Θέλει να βάλει μια τάξη στα διάφορα μέσα πληροφοριών που έχει στο σπίτι του και το γραφείο του (βιβλία, CD, DVD). Θέλει λοιπόν να του γράψετε μια εφαρμογή η οποία θα του επιτρέπει να καταχωρεί τι μέσα έχει και να μπορεί να τα αναζητά εύκολα, κάτι σαν μια προσωπική εφαρμογή ηλεκτρονικής βιβλιοθήκης ή βίντεο κλαμπ. Μάλιστα σας παραθέτει και τις ακόλουθες απαιτήσεις για την εφαρμογή _eLibrary_:

* Το πρόγραμμα θα πρέπει να αποθηκεύει βιβλία, μουσικά CD, βίντεο DVD και ψηφιακούς δίσκους δεδομένων.
* Ο χρήστης θα πρέπει να μπορεί να ενημερώνει ένα μέσο ακόμα και να το διαγράφει αν δεν το χρειάζεται πια.
* Ο χρήστης θα πρέπει να μπορεί να βλέπει μια πλήρη λίστα όλων των μέσων και να μπορεί να αναζητά ένα μέσο.
* Η αναζήτηση ενός βιβλίου θα μπορεί να γίνεται είτε μέσω ISBN, είτε μέσω τίτλου ή λέξεων κλειδιών
* Η αναζήτηση ψηφιακών θα μπορεί να γίνεται είτε μέσω τίτλου είτε μέσω λέξεων κλειδιών

Αφού κάνετε την ανάλυση των ανωτέρω απαιτήσεων, καταλήγετε στα σενάρια που φαίνονται στο διάγραμμα της εικόνας 1.

**Εικόνα 2** _Διάγραμμα κλάσεων της εφαρμογής eLibrary_

Σας δίνετε ήδη ένα μέρος της άσκησης. Όταν εκτελέσετε την κλάση ```Main``` θα σας εμφανιστεί το ακόλουθο μενού:

```
=========== MENU ===========
== 1. List media ===========
== 2. Search for a medium ==
== 3. Add new book =========
== 4. Add new disk =========
== 5. Edit medium ==========
== 6. Delete medium ========
== 0. Exit =================
============================
-> 4
====== 4. DISK MENU ========
== 1. Add new data disk ===
== 2. Add new Music CD ====
== 3. Add new Video DVD ===
== 0. Back ================
============================
```
Το πρόγραμμα δουλεύει ήδη για δίσκους δεδομένων (```Disk```) και ```CD``` και μένει να υλοποιήσετε την υποστήριξη ```DVD```. 

Στο βίντεο του μαθήματος θα δείτε πώς θα πρέπει να δουλεύει το πρόγραμμα.

Σας δίνονται οι παρακάτω κλάσεις:

* ```gr.cup.mathesis.elibrary.Main``` που υλοποιεί τη διεπαφή με το χρήστη
* ```gr.cup.mathesis.elibrary.model.library.ILibrary``` που είναι μια διεπαφή Java που υλοποιεί τις απαιτήσεις του χρήστη
* ```gr.cup.mathesis.model.ibrary.Library``` μια αρχική υλοποίηση της διεπαφής ```ILibrary``` που θα πρέπει να υλοποιήσετε κάποιες μεθόδους της ώστε να υποστηρίξετε ```DVD```. Η κλάση αυτή καλείται από την ```Main``` μέσω της στατικής μεθόδου ```getInstance()``` που επιτρέπει μόνο ένα αντικείμενο αυτής της κλάσης να δημιουργηθεί. Αυτό το pattern ονομάζεται _Singleton_.
* ```gr.cup.mathesis.elibrary.model.Medium``` που είναι η υπερκλάση όλων των μέσων που υποστηρίζονται από το πρόγραμμα
* ```gr.cup.mathesis.elibrary.model.Book``` που είναι κλάση βιβλίο
* ```gr.cup.mathesis.elibrary.model.Disk``` που είναι η κλάση ψηφιακού δίσκου δεδομένων, π.χ. προγράμματα
* ```gr.cup.mathesis.elibrary.model.CD``` που είναι η κλάση μουσικών CD
* ```gr.cup.mathesis.elibrary.model.DVD``` που είναι η κλάση βίντεο DVD που θα πρέπει να υλοποιήσετε σύμφωνα με την Εικόνα 3.

1. Πατήστε στο κουμπί φίλτρο και αφήστε την προεπιλογή ```TODOs only``` όπως φαίνεται στην παρακάτω εικόνα.
